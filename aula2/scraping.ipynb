#Extraindo preco de livros
com esse codigo vai ser extraido:
* Titulo do livro
* Preco do livro

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#URL do site
myurl= 'http://olhardigital.com.br/2024/09/12/pro/ia-teria-encontrado-forma-de-nao-ser-controlada-por-humanos-entenda/'
response= requests.get(myurl)
```

```python
response
```

```python
#Verificando se a requisicao foi bem-sucedida
if response.status_code == 200:
    html_content = response.text # Armazenando o conteudo html
    print ("Pagina acessada com sucesso")
else:
    print(f'Erro ao acessar a pagina:{response.status_code}')
```

```python
html_content
```

```python
#Criando objeto BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
#soup #objeto soup
```

```python
#1 Pegando o titulo da pagina
title = soup.title.string
print (f"**Titulo d pagina:**{title}")

print ("-------")
```

```python
#2 Econtrando todas as tagas de cabecalho
for i in range (1,7):
    headers = soup.find_all(f'h{i}')
    for header in headers:
        print (f'h{i}: {header.get_text(strip=True)}')
```

```python
#3 Extraindo todos os links

links = soup.find_all('a')
print(f"Total de links encontrados: {len(links)}\n")

# Exibindo apenas os primeiros 10 links
for link in links[:10]:
    href = link.get('href')
    text = link.get_text(strip=True)
    print(f"Texto: {text} | URL: {href}")
```
